```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'RSB_DataSet.csv'
data = pd.read_csv(file_path)

# Sort data by Score for better visualization
data_sorted = data.sort_values(by='Score', ascending=False)

# Bar chart
plt.figure(figsize=(15, 10))
sns.barplot(x='Score', y='Country', data=data_sorted, palette='viridis')
plt.title('Press Freedom Scores by Country')
plt.xlabel('Press Freedom Score')
plt.ylabel('Country')
plt.show()
```
