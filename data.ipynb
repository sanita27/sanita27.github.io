```python

# Load the CSV data
csv_url = 'https://raw.githubusercontent.com/sanita27/sanita27.github.io/main/merged_data_with_year.csv'
df = pd.read_csv(csv_url)

# Print the unique country names from the CSV
print(df['Country'].unique())

# Strip any leading/trailing whitespace and ensure consistent casing
df['Country'] = df['Country'].str.strip().str.title()

# Save the preprocessed CSV
preprocessed_csv_path = 'preprocessed_merged_data_with_year.csv'
df.to_csv(preprocessed_csv_path, index=False)
```

```python
import requests

# Load the GeoJSON data
geojson_url = 'https://raw.githubusercontent.com/sanita27/sanita27.github.io/main/countries.geojson'
geojson_data = requests.get(geojson_url).json()

# Extract country names
country_names_geojson = [feature['properties']['ADMIN'] for feature in geojson_data['features']]

# Print the unique country names
print(set(country_names_geojson))
```
